```python
# gerekli kütüphaneler import edilir.
import os
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam, Adagrad
from tensorflow.keras.models import load_model
import cv2 as cv
```

```python
# Veri seti klasörü ve sınıf etiketleri
dataset_folder = "WeatherDataset"
class_labels = ["Cloudy","Rain", "Shine","Sunrise"]

# Görüntülerin boyutu ve kanal sayısı
image_width, image_height = 150, 150
channels = 3

# Veri ve etiketleri depolamak için boş listeler
data = []
labels = []
```

```python
# Görüntüleri yükleme ve ön işleme
for label_idx, class_label in enumerate(class_labels): # her sınıf için for döngüsü oluşturuldu
    class_folder = os.path.join(dataset_folder, class_label) # sınıf klasörünün yolu oluşturuldu
    for image_filename in os.listdir(class_folder):
        image_path = os.path.join(class_folder, image_filename) # görüntü dosyasının yolu oluşturuldu 
        image = load_img(image_path, target_size=(image_width, image_height)) #görüntü belirtilen boyutlara göre yüklendi
        image_array = img_to_array(image) # görüntü diziye dönüştürüldü
        data.append(image_array) # data listesine eklendi
        labels.append(label_idx) # sınıf etiketlerine listesine eklendi
```

```python
# Verileri Numpy dizisine dönüştürme
data = np.array(data, dtype="float32")
labels = np.array(labels)

# Etiketleri kategorik hale getirme (One-Hot Encoding)
labels = to_categorical(labels)
```

```python
# Veri setini eğitim ve test kümelerine ayırma
from sklearn.model_selection import train_test_split

train_data, test_data, train_labels, test_labels = train_test_split(data, labels, test_size=0.3, random_state=42)
```

```python
# Modeli oluşturma
model = Sequential()

# Convolutional katmanı eklendi. Aktivasyon fonksiyonumuz relu ve verinin giriş bilgileri verildi.
model.add(Conv2D(32, (3, 3), activation="relu", input_shape=(image_width, image_height, channels)))

#2x2 boyutunda maksimum havuzlama katmanı. Görüntünün boyutunu yarıya indirir.
model.add(MaxPooling2D((2, 2)))

# Convolutional katmanı eklendi. Aktivasyon fonksiyonumuz relu.
model.add(Conv2D(64, (3, 3), activation="relu"))

# Bu kod, bir maksimum havuzlama katmanı daha ekler.
model.add(MaxPooling2D((2, 2)))

# bir başka evrişim katmanı ekler.
model.add(Conv2D(128, (3, 3), activation="relu"))

# Bu kod, bir maksimum havuzlama katmanı daha ekler.
model.add(MaxPooling2D((2, 2)))

# Bu kod, aşırı uyumun önlenmesi için %25 drop-out (atma) oranıyla bir drop-out katmanı ekler. 
model.add(Dropout(0.25))

# Bir önceki katmandan gelen veriyi tek boyutlu vektör haline çevirdik.
model.add(Flatten())

# Bu kod, 128 nörona sahip bir tam bağlantılı (dense) katman ekler. 'relu' aktivasyon fonksiyonu kullanılır.
model.add(Dense(128, activation="relu"))

# Bu kod, sınıf sayısına göre çıktı nöronlarını oluşturmak için bir başka tam bağlantılı katman ekler.
model.add(Dense(len(class_labels), activation="softmax"))
```

```python
# Modeli oluşturma
model3 = Sequential()

# Convolutional katmanı eklendi. Aktivasyon fonksiyonumuz relu ve verinin giriş bilgileri verildi.
model3.add(Conv2D(64, (3, 3), activation="elu", input_shape=(image_width, image_height, channels)))

#2x2 boyutunda maksimum havuzlama katmanı. Görüntünün boyutunu yarıya indirir.
model3.add(MaxPooling2D((2, 2)))

# Convolutional katmanı eklendi. Aktivasyon fonksiyonumuz relu ve verinin giriş bilgileri verildi.
model3.add(Conv2D(128, (3, 3), activation="sigmoid", input_shape=(image_width, image_height, channels)))

#2x2 boyutunda maksimum havuzlama katmanı. Görüntünün boyutunu yarıya indirir.
model3.add(MaxPooling2D((2, 2)))

# Bu kod, aşırı uyumun önlenmesi için %25 drop-out (atma) oranıyla bir drop-out katmanı ekler. 
model3.add(Dropout(0.5))

# Bir önceki katmandan gelen veriyi tek boyutlu vektör haline çevirdik.
model3.add(Flatten())

# Bu kod, sınıf sayısına göre çıktı nöronlarını oluşturmak için bir başka tam bağlantılı katman ekler.
model3.add(Dense(len(class_labels), activation="softmax"))
```
